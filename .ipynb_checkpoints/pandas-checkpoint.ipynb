### 3. How is the data distributed across different invoice dates? Are there any seasonal trends or patterns? (Use a graph for this.)

```python
#### The method to get the date was taken from ChatGPT
### The seasonal trends show that there is usually a spike in the sales during the start of the year (around January) and also during near the middle of the year (around May).
### The peak sales would be during the beginning of 2022, with the category of clothing bringing in the most sales.
```

```python
#How the data is dristributed across different invoice dates YEARLY.
#The findings from this graph indicates that there was a steady amount of sales during 2021. However, the sales started dipping from the start of 2022 and continued on its downward spiral to the end.

import pandas as pd
import matplotlib.pyplot as plt

merged_df['date_column'] = pd.to_datetime(merged_df['invoice_date'])

sales_by_year = merged_df.groupby(merged_df['date_column'].dt.year)['sales'].sum()

plt.plot(sales_by_year.index, sales_by_year.values)

plt.xlabel('Year')
plt.ylabel('Total Sales')
plt.title('Total Sales by Year')

plt.show()
```

```python
#How the data is dristributed across different invoice dates MONTHLY.
#FINDINGS:
#This provides a more micro-lens towards the different months for the past two years. 
#The significant pattern that has followed over the months wouold be significant spikes during the start of the years or the month of January.
#July also marks one of the biggest spikes during the mid-year season, with October slightly fluctuating the spectrum with a slight increase.

import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named merged_df with a 'invoice_date' and 'sales' column
merged_df['date_column'] = pd.to_datetime(merged_df['invoice_date'])

# Group the sales data by month and calculate the total sales for each month
sales_by_month = merged_df.groupby(pd.Grouper(key='date_column', freq='M'))['sales'].sum()

# Plot the total sales by month
plt.plot(sales_by_month.index, sales_by_month.values)

# Customize the plot
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Total Sales by Month')

plt.show()
```

```python
#How the data is dristributed across different ages with different categories.
#FINDINGS:
#The increase of sales may be contributed by certain events occurring in different age groups with the largest market seeming to be in their 30s to 50s.
#Considering this market, there must be an analysis conducted on events occurring within January and July (and perhaps October) to have a deeper understanding on what causes this seasonal pattern.

import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named merged_df with an 'age' column and 'sales' column

# Group the sales data by age and calculate the total sales for each age
sales_by_age = merged_df.groupby('age')['sales'].sum()

# Plot the total sales by age
plt.plot(sales_by_age.index, sales_by_age.values)

# Customize the plot
plt.xlabel('Age')
plt.ylabel('Total Sales')
plt.title('Total Sales by Age')

plt.show()
```

```python
#How the data is dristributed across different ages with different categories.
#FINDINGS
#With clothing being the most purchased category across all age groups, this may encourage the malls to provide more variety in their clothing options.
#This may also hint towards seasonal changes in terms of weather which may be the cause of the increased purchasing power of the consumers (e.g. Fall beginning in August in some countries, October marking the beginning of winter, etc.)

import pandas as pd
import matplotlib.pyplot as plt

sales_by_category_age = merged_df.groupby(['category', 'age'])['sales'].sum()

for category in merged_df['category'].unique():
    category_data = sales_by_category_age.loc[category]
    plt.plot(category_data.index, category_data.values, label=category)

plt.xlabel('Age')
plt.ylabel('Total Sales')
plt.title('Total Sales by Age, separated by Category')
plt.legend()

plt.show()
```

```python
#This shous how the sales for different categories vary through the months, adding the accumulated sales per month, with reference to the invoice dates.
#FINDINGS:
#It becomes clear here that the largest spike in sales occurs in January, followed by July for clothing, which means that this should be the main priority of the company to sell.
#Shoes had also been popular in January, and its second spike occurs around May, which means that they can increase shoe sales during this time to serve as greater incentives for people.
#For technology, the second spike after the initial January spike would be in October, so the marketing should be more directed towards technology during this time.

import pandas as pd
import matplotlib.pyplot as plt


merged_df['month'] = pd.to_datetime(merged_df['date_column']).dt.month

sales_by_category_month = merged_df.groupby(['category', 'month'])['sales'].sum()

for category in merged_df['category'].unique():
    category_data = sales_by_category_month.loc[category]
    plt.plot(category_data.index, category_data.values, label=category)

# Customize the plot
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.title('Total Sales by Month and Category')
plt.legend()

plt.show()
```
